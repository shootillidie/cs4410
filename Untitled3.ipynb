```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import matplotlib.pyplot as plt

from sklearn import model_selection
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import *
from sklearn.datasets import *
```

```python
!pip install scikit-learn # install the scikit-learn library that contains the load_breast_cancer function
from sklearn.datasets import load_breast_cancer # import the load_breast_cancer function from sklearn.datasets
import pandas as pd # import pandas for dataframe creation

data = load_breast_cancer()

df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target
df
```

```python
features = df.drop(columns = ['target'])
label = df['target']
```

```python
from sklearn.model_selection import train_test_split # import train_test_split from the model_selection module

X_train, X_test, y_train, y_test = train_test_split(features, label, test_size=0.2, random_state=42)
```

```python
# import the logistic regression algorithm
from sklearn.linear_model import LogisticRegression
```

```python
clf = LogisticRegression()
```

```python
#fit
clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)
#y_pred

y_pred_proba = clf.predict_proba(X_test)
#y_pred_proba

y_pred_binary = (y_pred_proba[:, 1] > 0.5).astype(int)
print(y_pred_proba)
print(y_pred_binary)
```

```python
# import the confusion_matrix function from the sklearn.metrics module
from sklearn.metrics import confusion_matrix

# import the ConfusionMatrixDisplay for visualization
from sklearn.metrics import ConfusionMatrixDisplay

# import matplotlib for plotting
import matplotlib.pyplot as plt

cm = confusion_matrix(y_test, y_pred, labels=clf.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=clf.classes_)
disp.plot()
plt.show()
```

```python
# import the classification_report function from the sklearn.metrics module
from sklearn.metrics import classification_report

report = classification_report(y_test, y_pred, target_names=['Malignant', 'Benign'])
print(report)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors = 5)
knn.fit(X_train, y_train)
knn_pred = knn.predict(X_test)

report = classification_report(y_test, knn_pred, target_names=['Malignant', 'Benign' ])
print(report)
```

```python
from sklearn.model_selection import KFold
kfold = KFold(n_splits=10, random_state=11, shuffle=True)
from sklearn.model_selection import cross_val_score
from sklearn.naive_bayes import GaussianNB # import the GaussianNB module
from sklearn.datasets import load_breast_cancer # import the load_breast_cancer function

nb = GaussianNB() # create the GaussianNB model object and assign it to nb
breast_cancer = load_breast_cancer() # load the breast cancer dataset

scores = cross_val_score(estimator=nb, X=breast_cancer.data,
                         y=breast_cancer.target, cv=kfold)
```
